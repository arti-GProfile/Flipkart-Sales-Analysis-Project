# Flipkart Sales Data Cleaning & Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
orders = pd.read_csv('../data/flipkart_raw_orders.csv')
products = pd.read_csv('../data/flipkart_products.csv')
customers = pd.read_csv('../data/flipkart_customers.csv')
```

```python
orders.drop_duplicates(inplace=True)
orders['Discount'].fillna(0, inplace=True)
orders['Order_Date'] = pd.to_datetime(orders['Order_Date'])
```

```python
sales = orders.merge(products,on='Product_ID').merge(customers,on='Customer_ID')
sales['Revenue'] = sales['Quantity'] * sales['Price'] * (1 - sales['Discount']/100)
sales.head()
```
